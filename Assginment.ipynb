```python
## construct a portfolio by using the first principle component based on SVD 
```

```python
# load data 
hist_df = pd.read_csv('data/dow.csv',index_col=0)
# convert price to log-ret;
ret_df = hist_df.apply(lambda x: np.log(x/x.shift()))[1:]
```

```python
from scipy import linalg
hist_data_win = 1000 # hist data to estimate
w_df = pd.DataFrame(0, index= ret_df.index[hist_data_win+1:], columns=ret_df.columns[:30]) # hold weight mat
# backtest strategy 
for t in range(hist_data_win,len(ret_df)-1):
    hdata = ret_df.iloc[t-hist_data_win:t,:30]
    cols_id = [ i for i in range(hdata.shape[1]) if not hdata.iloc[:,i].isnull().any() ]

    U,s,Vh = linalg.svd(hdata.iloc[:,cols_id]) # take svd 
    w = Vh.transpose()[:,0] # first comp portfolio weight
    w = w/s[0] # scale by variance 
    w = w/np.sum(w) # normalize weights 
    w_df.iloc[t-hist_data_win,cols_id] = w  
    
```

```python
port_ret = (ret_df.iloc[hist_data_win+1:,:]*w_df).sum(axis = 1) 
```

```python
res = pd.concat([port_ret, ret_df.iloc[:,-1]], axis=1,join='inner')
res.columns =['Port','DJI']
```

```python
res.head()
```

```python
res.cumsum().plot() # show cumulative return 
```

```python
# some portfolio stats 
total_return = lambda x: x.sum()
total_return.__name__= "total_return"

annual_return = lambda x : x.mean()*250
annual_return.__name__ = "annual_return"

volatility = lambda x : x.std()*np.sqrt(250)
volatility.__name__ = 'volatility'

sharpe = lambda x: x.mean()/x.std()*np.sqrt(250)
sharpe.__name__ = 'sharpe'


res.agg([total_return,annual_return,volatility,sharpe])
```

```python
#c) co-integration verify 
# if two series A,B are co-integrated, this means the difference between them are stationary. To make it more statistically 
# test easily, we want to check if the seires  log(A)-log(B) are stationary or not. 
import statsmodels.tsa.stattools as ts
import statsmodels.formula.api as sm
```

```python
fox_oil['log_cad'] = np.log(fox_oil['cad'])
fox_oil['log_oil'] = np.log(fox_oil['oil'])
```

```python
# fit a linear model 
fit = sm.ols(formula="log_cad ~ log_oil", data=fox_oil).fit()
fit.summary()
```

```python
plt.plot(fit.resid)
```

```python
# test if resid stationary
# Calculate and output the CADF test on the residuals
import pprint
cadf = ts.adfuller(fit.resid)
pprint.pprint(cadf)
```

```python
#d) pair trading 
```

```python
fox_oil.sort_index(inplace=True)
```

```python
fox_oil['spread'] = fox_oil['log_cad'] - fox_oil['log_oil']
```

```python
rolling_win = 100 # winddwo size 
fox_oil['spread_mean'] = fox_oil['spread'].rolling(rolling_win).mean()
fox_oil['spread_std'] = fox_oil['spread'].rolling(rolling_win).std()
fox_oil['plus_2std'] = fox_oil['spread_mean'] + 2 * fox_oil['spread_std']
fox_oil['minus_2std'] = fox_oil['spread_mean'] - 2 * fox_oil['spread_std']
```

```python
# trade rule 
fox_oil['trigger_signal'] =  np.nan 
fox_oil.loc[ (fox_oil.spread >= fox_oil.plus_2std) ,'trigger_signal'] = "trigger_plus"
fox_oil.loc[ (fox_oil.spread <= fox_oil.minus_2std) ,'trigger_signal'] = "trigger_minus"
fox_oil.loc[fox_oil.trigger_signal == fox_oil.trigger_signal.shift(1),'trigger_signal' ] = np.nan
```

```python
def rm_repeat(x):
    '''remove repeat element before another different elem'''
    last = x[0]
    for i in range(1,len(x)):
        if not pd.isnull(x[i]):
            if x[i] == last:
                x[i] = np.nan
            else:
                last = x[i]
    return x 
fox_oil['trigger_signal'] = rm_repeat(fox_oil.trigger_signal.values)
```

```python
fox_oil['order'] = np.nan 
fox_oil.loc[fox_oil.trigger_signal== 'trigger_plus','order'] = 'buy oil; sell cad' 
fox_oil.loc[fox_oil.trigger_signal== 'trigger_minus','order'] = 'sell oil; buy cad' 
```

```python
# show all the trades 
trade_df = fox_oil[~fox_oil.trigger_signal.isnull()]
```

```python
print('this strategy has {}  trades'.format(trade_df.shape[0]))
trade_df[['cad','oil','trigger_signal','order']]
```

```python
fox_oil['trigged_point'] = np.nan 
cond = pd.isnull(fox_oil.trigger_signal)
fox_oil.loc[~cond,'trigged_point'] = fox_oil.loc[~cond,'spread']
```
